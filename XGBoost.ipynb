```python
df = data
df.head()
```

```python
#using categorical dtypes
for i in data.columns[data.dtypes == 'object']:
    print(i,"::")
    df[i] = df[i].astype('category')
    df[i] = df[i].cat.codes
```

```python
#Feature variable
X = df.iloc[:,0:20]
X.head()
```

```python
#Setting the target variable
y = df['y']
y.head()
```

```python
#Splitting the dataset
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size = 0.8, test_size = 0.2, random_state = 0)
```

```python
X_train.shape
```

```python
#Building the model
#load the Gradient boosting from sklearn
from sklearn.ensemble import GradientBoostingClassifier
```

```python
GB_Classifier = GradientBoostingClassifier()
GB_Classifier.fit(X_train,y_train)
```

```python
y_predictions = GB_Classifier.predict(X_test)
```

```python
y_predictions
```

```python
#Load the accuracy score
from sklearn.metrics import accuracy_score
```

```python
print("Accuracy of the model::", accuracy_score(y_test,y_predictions))
```

```python
#Using Decision classifier
from sklearn.tree import DecisionTreeClassifier
```

```python
dTree = DecisionTreeClassifier()
```

```python
dTree.fit(X_train,y_train)
```

```python
y_pred = dTree.predict(X_test)
```

```python
#finding the accuracy of Decision Tree Classifier
print("Accuracy of the score::", accuracy_score(y_test,y_pred))
```

```python
#Using XGboost
XGBClass = xgb.XGBClassifier()
```

```python
XGBClass.fit(X_train,y_train)
```

```python
prediction = XGBClass.predict(X_test)
prediction
```

```python
print("Accuracy of the score::", accuracy_score(y_test,prediction))
```
